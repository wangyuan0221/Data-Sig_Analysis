```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
from matplotlib import cm
import os
import math
from scipy import optimize
from TWBP_Calculation import *
```

```python
# 広がる角度によって計算領域をコントロールする
RX_X = 100
RX_Y = 0
X_angle = 45
Y_angle = 45
height = 4800
length_x_half = height*np.tan(np.deg2rad(X_angle))
length_y_half = height*np.tan(np.deg2rad(Y_angle))
space_interval_x = WL/4
space_interval_y = height*np.tan(np.deg2rad(0.2))
grid_num_x = np.int64(np.round(length_x_half*2/space_interval_x))
grid_num_y = np.int64(np.round(length_y_half*2/space_interval_y))
capacity_x=np.linspace(-length_x_half,length_x_half,grid_num_x)
capacity_y=np.linspace(-length_y_half,length_y_half,grid_num_y)
# capacity_z = np.array([300,600,1200,2400,4800,9600])
```

```python
print("X軸格子数：{}".format(capacity_x.size))
print("Y軸格子数：{}".format(capacity_y.size))
print("格子総数：{}".format(capacity_x.size*capacity_y.size))
```

```python
result_twbp=np.load('../Signal/beam_broadening_check_twbp_NPI/x_1_4lamda_y_point2/twbp_4800m.npy')
power_original = np.abs(result_twbp*result_twbp.conjugate())
```

ビームパターン重み付き法

```python
# simple version
TX_Coordi = np.array([0,0,0])
RX_Coordi = np.array([RX_X,RX_Y,0])
Back_Wind = np.array([10,0,0]) # それぞれはu,v,w
def Doppler_Calcu(target_x,target_y,target_z):
    Target_Coordi=np.array([target_x,target_y,target_z])
    Unnormal_Vector_TX = Target_Coordi-TX_Coordi
    Unnormal_Vector_RX = Target_Coordi-RX_Coordi
    Unit_Vector_TX=1/np.linalg.norm(Unnormal_Vector_TX) * Unnormal_Vector_TX
    Unit_Vector_RX=1/np.linalg.norm(Unnormal_Vector_RX) * Unnormal_Vector_RX
    Unit_Vector = -(Unit_Vector_TX+Unit_Vector_RX)*1/2 # マイナスを取る意味：近づいてくる方向が正
    V_D = np.dot(Unit_Vector,Back_Wind)
    return V_D
Horizon_V_D = np.zeros_like(result_twbp,dtype=np.float128)
for idx_1 in range(capacity_x.size):
    for idx_2 in range(capacity_y.size):
        Horizon_V_D[idx_1][idx_2] = Doppler_Calcu(capacity_x[idx_1],capacity_y[idx_2],height)
```

```python
# ドップラー速度にビームパターンを重み付いた平均と標準偏差を計算する関数
def weighted_avg_and_std(values, weights):
    """
    Return the weighted average and standard deviation.

    values, weights -- Numpy ndarrays with the same shape.
    """
    average = np.average(values, weights=weights)
    # Fast and numerically precise:
    variance = np.average((values-average)**2, weights=weights)
    return (average, math.sqrt(variance))
res_weighted_avg_std=weighted_avg_and_std(Horizon_V_D,power_original)
print("重み付き平均:{}".format(res_weighted_avg_std[0]))# 重み付き平均
print("重み付き標準偏差:{}".format(res_weighted_avg_std[1])) # 重み付き標準偏差
```

観測モデル評価法

```python
lag_original =signal.correlation_lags(in1_len=result_twbp[:,0].size,in2_len=result_twbp[:,0].size)
lag = lag_original * (capacity_x[1]-capacity_x[0])
Eastward_Wind = 10
Southward_Wind = 0
Horizon_Wind=np.sqrt(Eastward_Wind**2+Southward_Wind**2)
lag_time = lag/Horizon_Wind
acf_all=np.zeros((lag_time.size),dtype=np.complex128)
for idx in range(capacity_y.size):
    acf_all+=signal.correlate(in1=result_twbp[:,idx],in2=result_twbp[:,idx])
acf_all_abs = np.abs(acf_all)

def func_gaussian(param,x):
    a,mu,sigma = param
    return a*np.exp(-(x-mu)**2/(2*sigma**2))
def func_residual(param,x,y):
    return func_gaussian(param,x)-y
# パワースペクトルをフィッティング
sampling_interval = np.max(lag_time)*2/acf_all_abs.size
fk = np.zeros_like(acf_all_abs)
freq = np.zeros_like(acf_all_abs)
v_dop = np.zeros_like(freq)
# ACFをフーリエ変換
fk = np.fft.fftshift(np.abs(np.fft.fft(acf_all_abs)))
freq = np.fft.fftshift(np.fft.fftfreq(acf_all_abs.size,sampling_interval))
v_dop = freq * WL/2
initial_param = np.array([1e5,0,-2])
temp_psd = optimize.leastsq(func_residual,x0=initial_param,args=(v_dop,fk))
ans_psd=temp_psd[0][2]
print("パワースペクトル幅：{}".format(ans_psd))
```

```python
# ACFのプロット
fs = 40
fig = plt.figure(figsize=(14,5),dpi=80)
ax = fig.add_subplot(1,1,1)
ax.set_title('ACF({}m)'.format(height),fontsize=fs)
ax.plot(lag_time,acf_all_abs,label='ACF')
ax.tick_params(labelsize="30")
ax.set_xlabel("lag_time(s)",fontsize=fs)
# ax.set_xlim(-100,100)
# ax.set_ylim(0,0.25e-7)
ax.legend(fontsize=20)
ax.grid()
plt.show()
```

```python
# パワースペクトルのフィッティング結果のプロット
gaus_x = np.linspace(-np.max(v_dop),np.max(v_dop),1000)
fig = plt.figure(figsize=(14,5),dpi=80)
ax = fig.add_subplot(1,1,1)
ax.set_title('Power Spectral({}m)'.format(height),fontsize=fs)
ax.plot(v_dop,fk,label='PSD')
ax.plot(gaus_x,func_gaussian(temp_psd[0],gaus_x),label='Gaussian Fitting')
ax.tick_params(labelsize="30")
ax.set_xlabel("Doppler Velocity(m/s)",fontsize=fs)
ax.legend(fontsize=20)
ax.grid()
plt.show()
```
