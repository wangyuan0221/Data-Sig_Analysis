```python
import numpy as np
import glob
import matplotlib.pyplot as plt
from scipy import fftpack
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
from scipy.optimize import leastsq
from scipy import optimize
import sys
from sympy.solvers import solve
from sympy import Symbol
```

```python
ff = 46.5*1e6
cc = 299792458
ll = cc/ff
```

```python
#
#load data files
#
files = glob.glob("./../Signal/Spherical_ScatterSimulation/filter_obs75_spt_2000/mainlobe_zenith00mainlobe_azimuth000_TXsubarray25_RXelements01_time06/*")
files = sorted(files)
for file in files:
    print(file)
```

```python
#
#データ数
#
data_leng = 2048
data = np.loadtxt(files[0], delimiter=",", max_rows=data_leng)
#
#配列の定義
#
n_subarray = int((data.shape[1]-1)/2) #受信アンテナ個数
time = data[:,0]
dt = time[1] - time[0]
sig = np.zeros((n_subarray,data_leng), dtype=complex)

#
#信号の読み込み
#
for file in files:
    data = np.loadtxt(file, delimiter=",", max_rows=data_leng)
    for i in range(n_subarray):
        sig[i,:] = sig[i,:] + data[:,i+1] + 1j*data[:,i+1+n_subarray]

fs = 20
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
ax.set_title('RX(0,0)', fontsize=fs)
ax.tick_params(labelsize=fs-5)
ax.plot(time, sig[i,:].real, label='real')
ax.plot(time, sig[i,:].imag, linestyle='--', label='imag')
ax.grid()
ax.set_xlabel('time [s]', fontsize=fs)
plt.show()
    
```

```python
#
#parameter
#
fs = 20
incoherent = 2**3
dleng = int(sig[0,:].size/incoherent)

#
#figure
#
fig = plt.figure(figsize=(12,8))
power = np.zeros(dleng, dtype=float)
for idx_sig in range(incoherent):
    power = power + np.abs(fftpack.fft(sig[0,dleng*idx_sig:dleng*(idx_sig+1)]))**2
power = fftpack.fftshift(power)
freq = fftpack.fftshift(fftpack.fftfreq(n=dleng, d=dt))
dvel = freq*ll/2
ax = fig.add_subplot(1,1,1)
ax.set_title('RX(0,0) Zenith:0 Azimuth:0', fontsize=fs)
ax.tick_params(labelsize=fs-5)
ax.plot(dvel, power, marker=".")
ax.axvline(0, linestyle='--', color='r')
ax.grid()
ax.set_xlabel("Doppler Velocity [m/s]", fontsize=fs)
plt.show()
```

```python
def func_gaussian(param,x):
    a,mu,sigma = param
    return a*np.exp(-(x-mu)**2/(2*sigma**2))
def func_residual(param,x,y):
    return func_gaussian(param,x)-y
```

```python
initial_param = np.array([1,0,1])
ans = optimize.leastsq(func_residual,x0=initial_param,args=(dvel,power))
print("mu:{}".format(ans[0][1]))
print("sigma:{}".format(ans[0][2]))
```

```python
Optimize_Param_Vertical=np.array(ans[0])
plt.figure(figsize=(14,6))
gaus_x = np.linspace(np.min(dvel),np.max(dvel),1000)
plt.title('Power Spectrum',fontsize = fs)
plt.plot(dvel,power,marker=".",label="power spectral")
plt.plot(gaus_x,func_gaussian(Optimize_Param_Vertical,gaus_x),label="estimated")
# plt.xlim(-10,10)
# plt.axvline(ans[0][1],color="red",linestyle="--")
plt.tick_params(labelsize="16")
plt.xlabel("Doppler Velocity",fontsize=fs)
plt.legend(fontsize=fs)
plt.grid()
plt.show()
```
