```python
class bpwm_rec():
    """
    beam pattern weighting method(rectangle)
    s_angle:solid angle(==1は立体角あり)
    ivl_x:x軸格子間隔設定値(interval)
    ivl_xreal:x軸格子間隔実際値
    """
    def __init__(self,x_angle,y_angle,s_angle,rx_x=0,rx_y=0,wind_ew=10,wind_ns=0,wind_ver=0,height=300,ivl_x=3,ivl_y=3):
        self.rx_x = rx_x
        self.rx_y = rx_y
        self.x_angle = x_angle
        self.y_angle = y_angle
        self.s_angle = s_angle
        self.wind_ew = wind_ew
        self.wind_ns = wind_ns
        self.wind_ver = wind_ver
        self.back_wind = np.array([self.wind_ew,self.wind_ns,self.wind_ver]) # それぞれはu,v,w
        self.height = height
        self.length_x_half = self.height*np.tan(np.deg2rad(self.x_angle))
        self.length_y_half = self.height*np.tan(np.deg2rad(self.y_angle))
        self.ivl_x = ivl_x
        self.ivl_y = ivl_y
        self.gridnum_x = np.int64(np.round(self.length_x_half*2/self.ivl_x))
        self.gridnum_y = np.int64(np.round(self.length_y_half*2/self.ivl_y))
        self.capacity_x = np.linspace(-self.length_x_half,self.length_x_half,self.gridnum_x)
        self.capacity_y = np.linspace(-self.length_y_half,self.length_y_half,self.gridnum_y)
        self.ivl_xreal = self.capacity_x[1]-self.capacity_x[0]
        self.ivl_yreal = self.capacity_y[1]-self.capacity_y[0]
    
    def print_func(self):
        print("X軸格子数：{}".format(self.capacity_x.size))
        print("Y軸格子数：{}".format(self.capacity_y.size))
        print("格子総数：{}".format(self.capacity_x.size*self.capacity_y.size))
        print("X軸格子間隔：{}m".format(self.ivl_xreal))
        print("Y軸格子間隔：{}m".format(self.ivl_yreal))
           
    def main_calcu(self):
        """
        res_bp:ビームパターンの計算結果
        s_angle:1は立体角を考慮、0は立体角を考慮しない
        res_bp:散乱電界強度
        power_ori:オリジナルパワー密度(W/m**2)
        power_nor:正規化されてるパワー密度
        power_db:デシベルパワー密度
        power:立体角をかけた後のパワー(W)
        """
        if (self.s_angle==1):
            self.res_bp = np.zeros((self.capacity_x.size-1,self.capacity_y.size-1),dtype=np.complex256)
            self.solid_angle = np.zeros_like(self.res_bp,dtype=np.float128)
            self.v_dop = np.zeros_like(self.res_bp,dtype=np.float128)
            for i in range(self.capacity_x.size-1):
                for j in range(self.capacity_y.size-1):
                    self.res_bp[i][j] = one_way_beampattern(self.capacity_x[i],self.capacity_y[j],\
                                                                  self.height,self.rx_x,self.rx_y,x_data,y_data)
                    theta = np.arctan(np.sqrt(self.capacity_x[i]**2+self.capacity_y[j]**2)/self.height)
                    path_tx_target = np.sqrt(self.capacity_x[i]**2+self.capacity_y[j]**2+self.height**2)
                    self.solid_angle[i][j] = self.ivl_xreal*self.ivl_yreal*np.cos(theta)/(path_tx_target-np.sin(theta)*self.ivl_xreal)**2
                    self.v_dop[i][j] = self.doppler_calcu_single(self.capacity_x[i],self.capacity_y[j],self.height)
            self.power_ori = np.abs(self.res_bp*self.res_bp.conjugate())
            self.power_nor = self.power_ori/np.max(self.power_ori)
            self.power_db = 10*np.log10(self.power_nor)
            self.power = self.power_ori*self.solid_angle
        elif (self.s_angle==0):
            self.res_bp = np.zeros((self.capacity_x.size,self.capacity_y.size),dtype=np.complex256)
            self.v_dop = np.zeros_like(self.res_bp,dtype=np.float128)
            for i in range(self.capacity_x.size):
                for j in range(self.capacity_y.size):
                    self.res_bp[i][j] = one_way_beampattern_rec(self.capacity_x[i],self.capacity_y[j],self.height,self.rx_x,self.rx_y,x_data,y_data)
                    self.v_dop[i][j] = self.doppler_calcu_single(self.capacity_x[i],self.capacity_y[j],self.height)
            self.power_ori = np.abs(self.res_bp*self.res_bp.conjugate())
            self.power_nor = self.power_ori/np.max(self.power_ori)
            self.power_db = 10*np.log10(self.power_nor)
        else:
            raise ValueError("s_angle must be 1 or 0")
    
    def plot_contourf(self,purpose,title,cbar_t):
        fs = 50
        x=self.capacity_x[:self.capacity_x.size-self.s_angle]
        y=self.capacity_y[:self.capacity_y.size-self.s_angle]
        xx,yy=np.meshgrid(x,y,indexing="ij")
        fig,ax=plt.subplots(constrained_layout=True,figsize=(20,20))
        fig_2D=ax.contourf(xx,yy,purpose,cmap="jet")
        ax.set_aspect('equal',adjustable='box')
        ax.set_title('{}'.format(title),fontsize=fs)
        ax.set_xlabel("Target_X(abscissa of scatters)(m)",fontsize=fs)
        ax.set_ylabel("Target_Y(ordinate of scatters)(m)",fontsize=fs)
        ax.tick_params(labelsize=fs-5)
        cbar=fig.colorbar(fig_2D,orientation="vertical")
        cbar.ax.set_ylabel("{}".format(cbar_t),fontsize=fs)
        cbar.ax.tick_params(labelsize=fs-5)
        plt.show()
    
    def check_solidangle(self):
        sum_solid_angle = np.sum(self.solid_angle)
        real_spherical_cap = 2*np.pi*(1-np.cos(np.deg2rad(self.x_angle)))
        print("格子領域ごとの立体角の総和：{}".format(sum_solid_angle))
        print("{}計算領域のspherical cap面積：{}".format(self.x_angle,real_spherical_cap))
        print("sum_solid/spherical cap:{}".format(sum_solid_angle/real_spherical_cap))
        print("4/3.14={}".format(4/np.pi))
    
    def doppler_calcu_single(self,target_x,target_y,target_z):
        RX_Coordi = np.array([self.rx_x,self.rx_y,0])
        Target_Coordi=np.array([target_x,target_y,target_z])
        Unnormal_Vector_TX = Target_Coordi
        Unnormal_Vector_RX = Target_Coordi-RX_Coordi
        Unit_Vector_TX=1/np.linalg.norm(Unnormal_Vector_TX) * Unnormal_Vector_TX
        Unit_Vector_RX=1/np.linalg.norm(Unnormal_Vector_RX) * Unnormal_Vector_RX
        Unit_Vector = -(Unit_Vector_TX+Unit_Vector_RX)*1/2 # マイナスを取る意味：近づいてくる方向が正
        V_D = np.dot(Unit_Vector,self.back_wind)
        return V_D

    def weighted_avg_and_std(self,values, weights):
        """
        Return the weighted average and standard deviation.

        values, weights -- Numpy ndarrays with the same shape.
        """
        average = np.average(values, weights=weights)
        # Fast and numerically precise:
        variance = np.average((values-average)**2, weights=weights)
#         print("重み付き平均:{}".format(average))# 重み付き平均
        print("{}度:{}".format(self.x_angle,math.sqrt(variance))) # 重み付き標準偏差
        return (average, math.sqrt(variance))

    
#球形領域
class bpwm_sph():
    """
    bpwm_sph:beam pattern weighting method(spherical)
    zenith_cgy:天頂角範囲(度数)
    """
    def __init__(self,zenith_cgy,s_angle,azimuth_cgy=360,rx_x=0,rx_y=0,radius_height=300,wind_ew=10,wind_ns=0,wind_ver=0):
        self.rx_x = rx_x
        self.rx_y = rx_y
        self.s_angle = s_angle
        self.radius_height = radius_height #球形の半径、高さでもある
        self.wind_ew = wind_ew
        self.wind_ns = wind_ns
        self.wind_ver = wind_ver
        self.back_wind = np.array([self.wind_ew,self.wind_ns,self.wind_ver]) # それぞれはu,v,w
        self.zenith_cgy = zenith_cgy
        self.zenith_radian = self.zenith_cgy*np.pi/180 #天頂角の範囲
        self.azimuth_cgy = azimuth_cgy
        self.azimuth_radian = self.azimuth_cgy*np.pi/180 #方位角の範囲
        self.azimuth_gridnum = int(360*(self.radius_height/300))
        self.zenith_gridnum = int(self.zenith_cgy*8*(self.radius_height/300))
        self.capacity_azimuth = np.linspace(0,self.azimuth_radian,self.azimuth_gridnum,endpoint=False) #緯度方向格子の方位角極座標
        self.capacity_zenith = np.linspace(self.zenith_radian,0,self.zenith_gridnum,endpoint=False)[::-1] #経度方向格子の天頂角極座標
        self.ivl_zenith_radian = self.capacity_zenith[1]-self.capacity_zenith[0] #経度方向格子間隔(弧)
        self.ivl_azimuth_radian = self.capacity_azimuth[1]-self.capacity_azimuth[0] #緯度方向格子間隔(一番下の円)(弧)
        # 天頂角と方位角により、直交座標系に変換する
        self.capacity_x = np.zeros((self.capacity_zenith.size,self.capacity_azimuth.size),dtype=np.float128) #直交座標系の初期化
        self.capacity_y = np.zeros_like(self.capacity_x)
        self.capacity_z = np.zeros_like(self.capacity_x)
        for i in range(self.capacity_zenith.size):
            t = self.capacity_zenith[i]
            for j in range(self.capacity_azimuth.size):
                p = self.capacity_azimuth[j]
                self.capacity_x[i,j] = self.radius_height * np.sin(t) * np.cos(p)
                self.capacity_y[i,j] = self.radius_height * np.sin(t) * np.sin(p)
                self.capacity_z[i,j] = self.radius_height * np.cos(t)
    
    def check_spherical(self):
        print("天頂角方向格子数：{}".format(self.zenith_gridnum))
        print("方位角方向格子数：{}".format(self.azimuth_gridnum))
        print("天頂角方向分割度数：{}".format(self.zenith_cgy/self.zenith_gridnum))
        print("方位角方向分割度数：{}".format(self.azimuth_cgy/self.azimuth_gridnum))
        print("格子総数：{}".format(self.zenith_gridnum*self.azimuth_gridnum))
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.plot_surface(self.capacity_x,self.capacity_y,self.capacity_z,color='red',antialiased=True,shade=True)
        plt.show()
        

    def main_calcu(self,sg_version):
        """
        sg_version:solid angleのバージョン、０はシンプル、１はコンプレックス(アンテナごとに考える)
        """
        if (self.s_angle==1):
            self.res_bp = np.zeros((self.capacity_zenith.size-1,self.capacity_azimuth.size),dtype=np.complex256)
            self.solid_angle = np.zeros_like(self.res_bp,dtype=np.float128)
            self.v_dop = np.zeros_like(self.res_bp,dtype=np.float128)
            for i in range(self.capacity_zenith.size-1):
                for j in range(self.capacity_azimuth.size):
                    self.res_bp[i][j]=one_way_beampattern(self.capacity_x[i][j],self.capacity_y[i][j],self.capacity_z[i][j],self.rx_x,self.rx_y,x_data,y_data)
                    self.power_weights = beampattern_weights(self.capacity_x[i][j],self.capacity_y[i][j],self.capacity_z[i][j],self.rx_x,self.rx_y,x_data,y_data)
                    if(sg_version==0):
                        self.solid_angle[i][j]=np.sin(self.capacity_zenith[i])*self.ivl_zenith_radian*self.ivl_azimuth_radian
                    elif(sg_version==1):
                        self.solid_angle[i][j] = np.average(self.sph_solid_angle_calcu_single(self.capacity_x[i][j],self.capacity_y[i][j],self.capacity_z[i][j],self.capacity_zenith[i])\
                                                                    ,weights=self.power_weights)
                    else:
                        raise ValueError("sg_version must be 1 or 0")
                    self.v_dop[i][j] = self.doppler_calcu_single(self.capacity_x[i][j],self.capacity_y[i][j],self.capacity_z[i][j])
            self.power_ori = np.abs(self.res_bp*self.res_bp.conjugate())
            self.power_nor = self.power_ori/np.max(self.power_ori)
            self.power_db = 10*np.log10(self.power_nor)
            self.power = self.power_ori*self.solid_angle
        elif (self.s_angle==0):
            self.res_bp = np.zeros((self.capacity_zenith.size,self.capacity_azimuth.size),dtype=np.complex256)
            self.v_dop = np.zeros_like(self.res_bp,dtype=np.float128)
            for i in range(self.capacity_zenith.size):
                for j in range(self.capacity_azimuth.size):
                    self.res_bp[i][j]=one_way_beampattern(self.capacity_x[i][j],self.capacity_y[i][j],self.capacity_z[i][j],self.rx_x,self.rx_y,x_data,y_data)
                    self.v_dop[i][j] = self.doppler_calcu_single(self.capacity_x[i][j],self.capacity_y[i][j],self.capacity_z[i][j])
            self.power_ori = np.abs(self.res_bp*self.res_bp.conjugate())
            self.power_nor = self.power_ori/np.max(self.power_ori)
            self.power_db = 10*np.log10(self.power_nor)
        else:
            raise ValueError("s_angle must be 1 or 0")
                    
    
    def sph_solid_angle_calcu_single(self,tar_x,tar_y,tar_z,theta):
        """
        sdag_single_ant:アンテナ一本にとっての立体角(sdag:solid angle)
        """
        sdag_single_ant = np.zeros((25,19),dtype=np.float128)
        for i in range(25):
            for j in range(19):
                path_ant_tar = np.sqrt((tar_x-x_data[i][j])**2+(tar_y-y_data[i][j])**2+tar_z**2) #o:original #p
                path_ant_o = np.sqrt(x_data[i][j]**2+y_data[i][j]**2) #d
                path_o_tar = np.sqrt(tar_x**2+tar_y**2+tar_z**2) #t
                cos_alhpa = (path_o_tar**2+path_ant_tar**2-path_ant_o**2)/(2*path_o_tar*path_ant_tar)
                sdag_single_ant[i][j] = (self.radius_height**2*np.sin(theta)*self.ivl_azimuth_radian*self.ivl_zenith_radian*cos_alhpa)/path_ant_tar**2
        return sdag_single_ant
    
    def doppler_calcu_single(self,target_x,target_y,target_z):
        RX_Coordi = np.array([self.rx_x,self.rx_y,0])
        Target_Coordi=np.array([target_x,target_y,target_z])
        Unnormal_Vector_TX = Target_Coordi
        Unnormal_Vector_RX = Target_Coordi-RX_Coordi
        Unit_Vector_TX=1/np.linalg.norm(Unnormal_Vector_TX) * Unnormal_Vector_TX
        Unit_Vector_RX=1/np.linalg.norm(Unnormal_Vector_RX) * Unnormal_Vector_RX
        Unit_Vector = -(Unit_Vector_TX+Unit_Vector_RX)*1/2 # マイナスを取る意味：近づいてくる方向が正
        V_D = np.dot(Unit_Vector,self.back_wind)
        return V_D
    
    def plot_contourf_sph(self,purpose,title,cbar_t):
        fs = 30
        fw = 1000
        x=self.capacity_azimuth
        y=self.capacity_zenith[:self.capacity_zenith.size-self.s_angle]*180/np.pi
        xx,yy=np.meshgrid(x,y,indexing="ij")
        fig = plt.figure(figsize=(20,20))
        ax = fig.add_subplot(111,polar=True)
        fig_2D=ax.contourf(xx,yy,purpose.T,cmap="jet")
        # ax.set_rmax(45)
        ax.set_title('{}'.format(title),fontsize=fs)
        ax.set_theta_zero_location('E')
        ax.tick_params(labelsize=fs-5)
        cbar=fig.colorbar(fig_2D,orientation="vertical")
        cbar.ax.set_ylabel("{}".format(cbar_t),fontsize=fs)
        cbar.ax.tick_params(labelsize=fs-5)
        plt.show()
        
    def weighted_avg_and_std(self,values, weights):
        """
        Return the weighted average and standard deviation.

        values, weights -- Numpy ndarrays with the same shape.
        """
        average = np.average(values, weights=weights)
        # Fast and numerically precise:
        variance = np.average((values-average)**2, weights=weights)
    #         print("重み付き平均:{}".format(average))# 重み付き平均
        print("{}度:{}".format(self.zenith_cgy,math.sqrt(variance))) # 重み付き標準偏差
        return (average, math.sqrt(variance))
```
