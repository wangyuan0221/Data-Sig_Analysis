```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# FigureAxes
fig = plt.figure(figsize = (12, 12))
ax = fig.add_subplot(111, projection='3d')
ax.grid()
ax.set_xlabel("Eastward Wind(m/s)", fontsize = 16)
ax.set_ylabel("Northward Wind(m/s)", fontsize = 16)
ax.set_zlabel("Vertical Wind(m/s)", fontsize = 16)
ax.set_xlim(0, 12)
ax.set_ylim(-12, 0)
ax.set_zlim(0, 3)

# O(3, 2, 4)
ax.quiver(0, 0, 0, 10, -10, 1,
          color = "red",
          arrow_length_ratio = 0.05)
```
